####Ques 2: Checking if the player have played for icc or not

```python
#finding the answer through pandas apply method
def icc(x):
  if 'ICC/' in x:
    return 1
  elif '/ICC' in x:
    return 1
  else:
    return 0
df['Played_for_ICC'] = df['Country'].apply(icc)
#display(df)
  
```

####Question 6: find the player based on who have given highest run,total wicket taken,total average, strike_rate etc.

```python
def top(x):
  #sorted the values according to query
  df_top = df.sort_values(by= x, ascending= False)
  #reseting the index after sorting
  df_top = df_top.reset_index()
  #removing the redundant column
  df_top = df_top.drop('index', axis= 1)
  #getting the player name by get values
  name_of_the_player = df_top._get_value(0, 'Player_name')
  print('name of the player who have highest ' + x +' is = ', name_of_the_player)
```

```python
#creating object of top() function
name = top('Total_Runs')
name = top('Average')
```

```python
def Player(x):
  df_top_player = df.nlargest(7, x)
  return df_top_player
```

```python
df_strikeRate = Player('StrikeRate')
df_average = Player('Average')
df_TotalRun = Player('Total_Runs')
df_wickets = Player('Wickets')
df_economy = Player('Economy')
df_years_played = Player('years_played')
```

```python

def plots(key, DATA):

  sns.barplot(x = 'Player_name', y = key , data = DATA)
  plt.xlabel('Name of the Player')
  plt.xticks(rotation = 90)
  plt.ylabel(key)
  plt.yticks()
  plt.title('Top 7 players based on ' + key)
```

```python
plt.figure(figsize= (20,14))


plt.subplot(2,3,1)
plot1 = plots('StrikeRate', df_strikeRate)
plt.tight_layout()

plt.subplot(2,3,2)
plot2 = plots('Average', df_average)
plt.tight_layout()

plt.subplot(2,3,3)
plot3 = plots('Total_Runs', df_TotalRun)
plt.tight_layout()

plt.subplot(2,3,4)
plot4 = plots('Wickets', df_wickets)
plt.tight_layout()

plt.subplot(2,3,5)
plot5 = plots('Economy', df_economy)
plt.tight_layout()

plt.subplot(2,3,6)
plot6 = plots('years_played', df_years_played)
plt.tight_layout()

plt.suptitle('Top Player Based on different parameters', x = 0.5, y = 1.035)
plt.tight_layout()
plt.show()
```
